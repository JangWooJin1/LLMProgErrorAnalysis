문제별로 학생들이 자주 틀리는 유형 분석

```python
import openai
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def question(query):

    # FTA API
    API_KEY = ""
    openai.api_key = API_KEY

    model = "gpt-3.5-turbo"  

    messages = [
        {"role":"system", "content":"You are a helpful assistant."},
        {"role":"user","content":query}
    ]

    response = openai.ChatCompletion.create(
        temperature=0.8,
        model=model,
        messages = messages
    )

    answer = response['choices'][0]['message']['content']

    return answer
```

```python
df = pd.read_excel("../3차결과.xlsx") 
result_df = df

df
```

```python
df = df.drop(["문제 이름","오류 분류","대분류", "제출 시간", "코드" , "대분류", "prediction"], axis=1)

print(df.to_string(index=False))
```

```python
# 학생별 정답 수 계산
correct_counts = df[df['정답 여부'] == 'AC'].groupby('학생명').size()

# 학생별 전체 문제수 계산
total_counts = df.groupby('학생명').size()

# 학생별 정답률 계산
correct_rates = correct_counts / total_counts

# 정답률과 함께 학생명을 포함한 새로운 데이터프레임 생성
result_df = pd.DataFrame({'학생명': correct_rates.index, '정답률': correct_rates.values})

result_df
```

```python
result_df['그룹'] = pd.cut(result_df['정답률'], 
                         bins=[0, 0.35, 0.65, 1], 
                         labels=['group3', 'group2', 'group1']
                    )

result_df
```

```python
group_counts = result_df.groupby('그룹').size()

group_counts
```
