```python
import openai
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def question(query):

    # FTA API
    API_KEY = ""
    openai.api_key = API_KEY

    model = "gpt-3.5-turbo"  

    messages = [
        {"role":"system", "content":"You are a helpful assistant."},
        {"role":"user","content":query}
    ]

    response = openai.ChatCompletion.create(
        temperature=0.8,
        model=model,
        messages = messages
    )

    answer = response['choices'][0]['message']['content']

    return answer
```

```python
# CSV 파일 읽기
df = pd.read_excel("실제 학생들의 제출 코드(통합).xlsx") 
df = df.drop(["오류 분류", "대분류", "prediction"], axis=1)
df = df.sort_values('제출 시간')
grouped = df.groupby("문제 이름")
group_names = grouped.groups.keys()
print(group_names)
```

```python
# 그룹화된 결과 출력
for name, group in grouped:
    print(name) 
    print(group.shape) 
```

```python
# input 
group_1_name = list(group_names)[0]
group_1 = grouped.get_group(group_1_name)
group_1 = group_1.drop(["문제 이름", "제출 시간"], axis=1)

question_query = f"""
1. Question :
The following Dataframe represents the students' problem-solving codes for {group_1_name} problem. 
Your task is to perform a code review and identify patterns in the codes. Please provide a code review and a pattern for each code.

a. Review each code and provide feedback or suggestions for improvement.
b. Identify a pattern for each code, categorizing it based on the type of error or issue present in the code.
c. Present the code reviews and patterns in a format that can be added to the DataFrame
"""


result_query = """
2. respond :
Please respond in the format below so that pattern and code reviews can be added to the Dataframe

create a pattern for each code
ex) 
Pattern Explanation: 
0 - Correct answer
1 - Compilation error
2 - ~~~~
etc

Pattern = [2, 0, 0,...]

provide code reviews in the following list format for each code:
ex) 
Code review = [ "~~~~", "~~~~" ...]

"""

context_query = """
3. Context:
The example of the correct answer code is as follows. Both the input format and output format should be identical for it to be recognized as the correct answer.
#include <cstdio>

int main()
{
int a;
scanf("%d", &a);

if(a < 10)
{printf("hello");}
}

"""

context2_query = "The DataFrame contains the following data.\n\n"  + group_1[:20].to_string(index=False)

query1_1 = question_query  + result_query + context_query + context2_query 

print(query1_1)
```

```python
ans1_1 = question(query1_1)
print(ans1_1)
```
